```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Salary.csv")
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
ar=si.fit_transform(dataset[["Age", "Experience", "Salary"]])
```

```python
dataset=pd.DataFrame(ar, columns=dataset.select_dtypes(include=["float64"]).columns)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="Age", y="Experience", data=dataset, hue="Salary")
plt.show()
```

```python
x=dataset.iloc[:, :-1]
y=dataset["Salary"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test), knn.score(x_train, y_train)
```

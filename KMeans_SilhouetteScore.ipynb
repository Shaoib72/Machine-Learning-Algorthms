```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("iris.csv")
```

```python
dataset.head(4)
```

```python
dataset.drop(columns=["species"], inplace=True)
```

```python
dataset.head(4)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2, 21):
    km=KMeans(n_clusters=i, init="k-means++")
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
km.labels_
```

```python
plt.figure(figsize=(15, 7))
plt.plot([i for i in range(2, 21)], wcss, marker='o')
plt.xlabel("no of clusters")
plt.xticks([i for i in range(2, 21)])
plt.ylabel("wcss")
plt.grid(axis="x")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmn=KMeans(n_clusters=2)
dataset['Predict']=kmn.fit_predict(dataset)
```

```python
kmn.labels_
```

```python
silhouette_score(dataset,labels=kmn.labels_ )
```

```python
dataset.head(4)
```

```python
ss=[] 
no_clus=[j for j in range(2, 21)]
for i in range(2, 21):
    km1=KMeans(n_clusters=i)
    km1.fit(dataset)
    ss.append(silhouette_score(dataset, km1.labels_))
```

```python
plt.figure(figsize=(15, 7))
plt.plot(no_clus, ss)
plt.xlabel("no of clusters")
plt.xticks(no_clus)
plt.ylabel("Silhouette_score")
plt.grid(axis="x")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Salary.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include=["float64"]).columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
```

```python
ar=si.fit_transform(dataset[['Age', 'Experience', 'Salary']])
```

```python
dataset=(pd.DataFrame(ar, columns=dataset.select_dtypes(include=["float64"]).columns))
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.heatmap(data=dataset.corr(), annot=True)
plt.show()
```

```python
x=dataset.iloc[:, :-1]
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.22, random_state=85)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
lr.predict([[32,5]])
```

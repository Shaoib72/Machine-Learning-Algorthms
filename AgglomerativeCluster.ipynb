```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("iris.csv")
dataset.head(4)
```

```python
dataset.drop(columns=["species"], inplace=True)
dataset.head(4)
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(30,30))
sc.dendrogram(sc.linkage(dataset, method='single', metric='euclidean'))
plt.savefig('dendro.jpg')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=3, linkage='ward')
dataset['Predict']=ac.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset, hue='Predict')
plt.show()
```
